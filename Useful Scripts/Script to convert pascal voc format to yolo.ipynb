```python
def convert(size, box):
    dw = 1./(size[0])
    dh = 1./(size[1])
    x = (box[0] + box[1])/2.0 - 1
    y = (box[2] + box[3])/2.0 - 1
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)
```

```python
#script to organize list in ascending order
#
read_xml_list= [os.path.join('/home/saadi/Desktop/test_xml _write/annotations', file) for file in os.listdir('/home/saadi/Desktop/test_xml _write/annotations') if file.endswith('.xml')]
print(read_xml_list)
read_xml_list.sort(key=lambda f: int(filter(str.isdigit, f))) # reorganizes the list
print('###################################################')
print(read_xml_list)
```

```python
write_yolo_label= [os.path.join('/home/saadi/Desktop/test_xml _write/yolo_labels', file) for file in os.listdir('/home/saadi/Desktop/test_xml _write/yolo_labels') if file.endswith('.txt')]
print(write_yolo_label)
write_yolo_label.sort(key=lambda f: int(filter(str.isdigit, f))) # reorganizes the list
print('###################################################')
print(write_yolo_label)
```

```python
for xml_read, yolo_write in zip(read_xml_list,write_yolo_label):
    print('The xml file is',xml_read)
    print('The yolo label write file is',yolo_write)
    in_file = open(xml_read,'r')
    out_file = open(yolo_write, 'w')
    tree=ET.parse(in_file)
    root = tree.getroot()
    size = root.find('size')
    w = int(size.find('width').text)
    h = int(size.find('height').text)

    for obj in root.iter('object'):
        difficult = obj.find('difficult').text
        cls = obj.find('name').text
        if cls not in classes :# or int(difficult)==1 #i removed this to add difficult
            continue
        cls_id = classes.index(cls)
        xmlbox = obj.find('bndbox')
        b = (float(xmlbox.find('xmin').text), float(xmlbox.find('xmax').text), float(xmlbox.find('ymin').text), float(xmlbox.find('ymax').text))
        bb = convert((w,h), b)
        out_file.write(str(cls_id) + " " + " ".join([str(a) for a in bb]) + '\n')
```
