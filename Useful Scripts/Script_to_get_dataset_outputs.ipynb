```python
import os
from PIL import Image

def get_image_sizes(root_folder):
    image_sizes = {}
    
    for root, dirs, files in os.walk(root_folder):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
                file_path = os.path.join(root, file)
                with Image.open(file_path) as img:
                    width, height = img.size
                    image_sizes[file] = (width, height)
    
    return image_sizes

    
root_folder = '/media/saadi/Drive/Thesis_Folder/Thesis_2022/Jpeg_organized'
image_sizes = get_image_sizes(root_folder)
for image, size in image_sizes.items():
    print(f"Image: {image}, Size: {size[0]}x{size[1]} pixels")
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Extracting width and height data for plotting
widths = [size[0] for size in image_sizes.values()]
heights = [size[1] for size in image_sizes.values()]

# Plotting histogram for image widths
plt.hist(widths, bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Image Widths')
plt.xlabel('Width')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plotting histogram for image heights
plt.hist(heights, bins=20, color='green', alpha=0.7)
plt.title('Distribution of Image Heights')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
